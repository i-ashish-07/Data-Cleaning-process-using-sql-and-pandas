## Email validation

```python
def add_missing_at(email):
    if isinstance(email, str) and '@' not in email:
        if 'gmail.com' in email:
            return email.replace('gmail.com', '@gmail.com')
        elif 'yahoo.com' in email:
            return email.replace('yahoo.com', '@yahoo.com')
        elif 'outlook.com' in email:
            return email.replace('outlook.com','@outlook.com')
    return email

df_clean['email'] = df['email'].apply(add_missing_at)
```
